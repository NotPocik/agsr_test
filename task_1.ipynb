```python
import pandas as pd
import csv
```

### Обработка auto.csv для корректного чтения

```python
def clean_csv_line(line):
    line = line.strip()
    if line.startswith('"') and line.endswith('"'):
        line = line[1:-1]
    line = line.replace('""', '"')
    return line

with open('auto.csv', 'r') as f:
    lines = [clean_csv_line(line) for line in f]

with open('auto_fixed.csv', 'w', encoding='utf-8') as file:
    file.write('\n'.join(lines))
```
